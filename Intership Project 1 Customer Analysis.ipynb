```python
OBJ = (Customer_df.dtypes == 'object')
OBJ_NUM = list(OBJ[OBJ].index)
print("Objects : ",len(OBJ_NUM))

INT= (Customer_df.dtypes == 'int64')
INT_NUM = list(INT[INT].index)
print("Integer : ",len(INT_NUM))

FLOAT = (Customer_df.dtypes == 'float64')
FLOAT_NUM = list(FLOAT[FLOAT].index)
print("Float : ",len(FLOAT_NUM))
```

```python
Unique = []
for col in OBJ_NUM:
    Unique.append(Customer_df[col].unique().size)
plt.figure(figsize=(10,6))
plt.title('Unique Value Identification')
plt.xticks(rotation=90)
sns.barplot(x=OBJ_NUM,y=Unique)
```

```python
plt.figure(figsize=(18, 36))
plt.title('Category Count')
plt.xticks(rotation=90)
index = 1

for col in OBJ_NUM:
    y = Customer_df[col].value_counts()
    plt.subplot(11, 4, index)
    plt.xticks(rotation=90)
    sns.barplot(x=list(y.index), y=y)
    index += 1
```

```python
droped_coloumns=['Year_Birth','Education','Marital_Status','Kidhome','Teenhome','Recency']
df=Customer_df.drop(columns=droped_coloumns,axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder

X = (df.dtypes == 'object')
OBJ_NUM = list(X[X].index)
print("Objects : ")
print(OBJ_NUM)
print('Number of Features : ',
   len(OBJ_NUM))
```

```python
Encoder = OneHotEncoder(sparse_output=False)
Cols = pd.DataFrame(Encoder.fit_transform(df[OBJ_NUM]))
Cols.index = df.index
Cols.columns = Encoder.get_feature_names_out()
Output_df = df.drop(OBJ_NUM, axis=1)
Output_df = pd.concat([Output_df, Cols], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()


scaled_Output = scaler.fit_transform(Output_df)
scaled_Output
```

```python
Result_df=pd.DataFrame(scaled_Output)
Result_df.columns=Output_df.columns
Result_df.head(10)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
 
X = Result_df.drop(['MntMeatProducts'], axis=1)
Y = Result_df['MntMeatProducts']
 

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, train_size=0.7, test_size=0.3, random_state=42)
```

```python
Linear_Model = LinearRegression()
Linear_Model.fit(X_train, Y_train)
Y_pred = Linear_Model.predict(X_test)
```

```python
import numpy as np
Y=np.sqrt(mean_squared_error(Y_test, Y_pred))
Y
```
